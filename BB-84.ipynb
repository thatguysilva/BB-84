```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(1, 1)
```

```python
qc.draw()
```

```python
qc.measure(0, 0)
```

```python
display(qc.draw(output='mpl'))
```

```python
from qiskit import Aer, execute
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

simulator = Aer.get_backend('qasm_simulator')
result = execute(qc, simulator, shots=1024).result()

# Get the histogram data
counts = result.get_counts(qc)

# Plot the histogram
plot_histogram(counts)
```

```python
from qiskit import IBMQ, execute
from qiskit.visualization import plot_histogram
from qiskit.providers.ibmq import least_busy

IBMQ.save_account('yourtokenhere')
IBMQ.load_account()


# Checking the backend and choosing the least busy

provider = IBMQ.get_provider(hub='ibm-q')

backend = least_busy(provider.backends(filters=lambda x: x.configuration().n_qubits >= 5
                                       and not x.configuration().simulator
                                       and x.status().operational==True))

print("We'll run on:", backend)
```

```python
result = execute(qc3, backend, shots=1024).result()

# Get the histogram data
counts = result.get_counts(qc3)

# Plot the histogram
plot_histogram(counts)
```

```python
qcf01 = QuantumCircuit(5, 5)
qcf01.draw("mpl")
```

```python
qcf01.x(0)
qcf01.h(0)
qcf01.h(1)
qcf01.x(3)
qcf01.h(4)
qcf01.draw("mpl")
```

```python
qcf01.barrier()
qcf01.draw("mpl")
```

```python
qcf01.i(0)
qcf01.h(1)
qcf01.h(2)
qcf01.i(3)
qcf01.i(4)
qcf01.draw("mpl")
```

```python
qcf01.measure(0, 0)
qcf01.measure(1, 1)
qcf01.measure(2, 2)
qcf01.measure(3, 3)
qcf01.measure(4, 4)
qcf01.draw("mpl")
```

```python
#SIMULATOR

simulator = Aer.get_backend('qasm_simulator')
result = execute(qcf01, simulator, shots=1024).result()

# Get the histogram data
counts = result.get_counts(qcf01)

# Plot the histogram
plot_histogram(counts)
```
